```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import sklearn
import pandas as pd
import os
import sys
import time
import tensorflow as tf
from tensorflow import keras

print(tf.__version__)
print(sys.version_info)
for module in mpl, np, pd, sklearn, tf, keras:
    print(module.__name__, module.__version__)
```

```python
fashion_mnist=keras.datasets.fashion_mnist
(x_train_all,y_train_all),(x_test,y_test)=fashion_mnist.load_data()
x_valid,x_train=x_train_all[:5000],x_train_all[5000:]
y_valid,y_train=y_train_all[:5000],y_train_all[5000:]

print(x_valid.shape,y_valid.shape)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
def make_dataset(data,label,epoch=10,shuffle=True,batch_size=32):
    dataset=tf.data.Dataset.from_tensor_slices((data,label))
    if shuffle:
        dataset=dataset.shuffle(10000)
    dataset=dataset.repeat(epoch).batch(batch_size)
    return dataset

train_dataset=make_dataset(x_train, y_train)
valid_dataset=make_dataset(x_valid, y_valid)
test_dataset=make_dataset(x_test, y_test)
# for item in train_dataset.take(1):
#     print(item)
    
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = [28,28]),
    keras.layers.Dense(300,activation = 'relu'),
    keras.layers.Dense(100,activation = 'relu'),
    keras.layers.Dense(10,activation = 'softmax')
])

adam=tf.keras.optimizers.Adam(lr=1e-5, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = adam,metrics = ['accuracy'])

model.layers
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model, to_file='./output/structure/model_test.png',show_shapes=True)
```

```python
logdir="./output/tensorboard/"
output_model_file="./output/model/h5/model.h5"

callbacks=[
    keras.callbacks.TensorBoard(logdir),
    keras.callbacks.ModelCheckpoint(output_model_file,save_best_only=True, save_weight_only=False),
    keras.callbacks.EarlyStopping(patience=5,min_delta=1e-3)
]

history = model.fit(train_dataset,epochs=10,validation_data=valid_dataset, callbacks=callbacks,verbose=1)
print(history.history)


def plot_learning_curves(history):
    pd.DataFrame(history.history).plot(figsize = (8,5))
    plt.grid(True)
    plt.gca().set_ylim(0,1)
    plt.show()

plot_learning_curves(history)
```

```python
graph_model_dir="./output/model/saved_graph"
tf.saved_model.save(model, graph_model_dir)
```

```python
graph_model = tf.saved_model.load(graph_model_dir)
print(graph_model.signatures.keys())
inference = graph_model.signatures["serving_default"]
```
